```python
bi=pd.read_csv('/home/asanchorf6/PythonMaster/RetoAmadeus/bookings.csv.bz2', sep='^', 
               usecols=['arr_port','pax','month'], chunksize=1000000)

all_chunks=pd.DataFrame()

for i,b in enumerate(bi):
    print(i, len(b))
    b.dropna()
    b['arr_port'] = b['arr_port'].str.strip()
    result_chunk=b[b['arr_port'].isin(['MAD', 'BCN', 'AGP'])].groupby(['arr_port', 'month']).sum()
    
    all_chunks=all_chunks.append(result_chunk)

result_all=all_chunks.reset_index().groupby(['arr_port', 'month']).sum().reset_index()
result_all
```

```python
result_all.shape
```

```python
fig, ax = plt.subplots(figsize=(20, 6))

mad = result_all[result_all.arr_port == 'MAD']
bcn = result_all[result_all.arr_port == 'BCN']
agp = result_all[result_all.arr_port == 'AGP']

ax.plot(mad.month, mad.pax, label='Madrid')
ax.plot(bcn.month, bcn.pax, label='Barcelona')
ax.plot(agp.month, agp.pax, label='Málaga')

ax.legend()

plt.show()
```

```python
result_all.to_csv('cities_by_month.csv', sep='^',index=False)
```
